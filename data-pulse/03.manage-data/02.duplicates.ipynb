# Duplicates

```python
# DUPLICADOS EXACTOS
datos_exactos = [
    {"id": 1, "nombre": "Juan", "email": "juan@example.com"},
    {"id": 2, "nombre": "María", "email": "maria@example.com"},
    {"id": 1, "nombre": "Juan", "email": "juan@example.com"},  # Duplicado exacto
    {"id": 3, "nombre": "Pedro", "email": "pedro@example.com"}
]
```

## Find exact duplicates

```python
# find exact duplicates
def find_exact_duplicates(items):
    visited = []
    duplicates = []
    
    for i in items:
        print("\n")
        print(f"{i.items()}")
        print(f"{tuple(i.items())}")
        print(f"{sorted(tuple(i.items()))}")
        i_tuple = sorted(tuple(i.items()))
        if i_tuple in visited:
            duplicates.append(i_tuple)
        else:
            visited.append(i_tuple)

    print(f"duplicates: {duplicates}")
        
find_exact_duplicates(datos_exactos)
```

## Find partial duplicates

```python
# DUPLICADOS PARCIALES (mismo ID, datos diferentes)
datos_parciales = [
    {"id": 1, "nombre": "Juan", "email": "juan@example.com"},
    {"id": 1, "nombre": "Juan", "email": "juan.nuevo@example.com"},  # Mismo ID, email diferente
    {"id": 2, "nombre": "María", "email": "maria@example.com"},
    {"id": 2, "nombre": "Maria", "email": "maria@example.com"}  # Mismo ID, nombre con tilde
]

# by filed
def find_by_filed_name(items: [], field_name: str):
    visited = {}
    duplicated = []
    
    for id, item in enumerate(items):
        print(f"{id}, field_name: {field_name} - {item.get(field_name)}")
        value = item.get(field_name)
        if value in visited:
            duplicated.append((visited[id], id, item))
        else:
            visited[value] = id
    
    print(f"visited: {visited}")
    print(f"duplicated: {duplicated}")
        

find_by_filed_name(datos_parciales, "id")
```
